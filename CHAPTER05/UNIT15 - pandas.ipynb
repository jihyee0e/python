```python
#예제) 전체 행에 대해서 A열의 값을 B열의 값으로 나눈 후, 그 결과를 새로 만든 D에 저장
import numpy as np
import pandas as pd

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index=index, columns=list('ABC'))

df['D'] = df['A'] / df['B']
df
```

```python
#예제) df 데이터프레임에서 행 우선으로 합 구하고, 그 결과를 E라는 열을 생성하여 저장

df['E'] = np.sum(df, axis=1)
df
```
