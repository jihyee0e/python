```python
import pandas as pd

df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table')
print(df)
```

```python
#인덱스를 이용하여 원하는 표를 개별 데이터프레임으로 지정
df[1]
```

```python
#헤더를 열 이름으로 지정, 나라 이름을 인덱스 이름으로 지정하는 속성 추가
import pandas as pd

##열, 행의 0번 라인 삭제
df = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table',
                 header=0, index_col=0)
df[1]
```

```python
#하계올림픽에 대한 데이터만 추출하기
summer = df[1].iloc[1:, :5]
summer
```

```python
#컬럼(열) 이름 수정하기
summer.columns = ['경기수', '금', '은', '동', '계']
summer
```

```python
#컬럼 정렬 예제)
import pandas as pd

df = pd.DataFrame([
    [16,7,10],
    [15,8,11],
    [14,9,12]],
    columns = ['a','b','c'],
    index = [1,2,3]
)
df
##(1) 컬럼을 기준으로 데이터프레임 정렬 - 'a'기준
df.sort_values('a')
##(2) 컬럼을 기준으로 시리즈 정렬 - 'a' 기준
df['a'].sort_values()
##(3) 정렬차순(내림차순)
df.sort_values('a', ascending=False)
##(4) 여러 컬럼을 기준으로 
df.sort_values(['b','c'], ascending=False)
```

```python
#금메달 기준으로 정렬하기
import pandas as pd

##데이터 불러오기
data = pd.read_html('https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table',
                   header=0, index_col=0)
##데이터 인덱스 1번의 행은 1번부터 끝, 열은 처음부터 4까지
summer = data[1].iloc[1:, :5]
##칼럼(열) 이름 수정하기
summer.columns = ['경기수', '금', '은', '동', '계']
##'금'을 기준으로 내림차순 정렬
summer.sort_values('금', ascending=False, inplace=True)
summer
```

```python
#엑셀 파일로 저장
summer.to_excel('하계올림픽메달.xlsx')
```
