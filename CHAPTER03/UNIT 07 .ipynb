- 내 생일의 최고 기온 및 최저 기온 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)  #>>['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']

high_result = []  #최고 기온 데이터
low_result = []  #최저 기온 데이터

for row in data:
    if row[-1] != '' and row[-2] != '':  #누락 데이터가 아니라면
        if 1983 <= int(row[0].split('-')[0]):  #1983년 이후 데이터만 추출
            if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '06':  #내 생일
                high_result.append(float(row[-1]))
                low_result.append(float(row[-2]))

plt.rcParams['axes.unicode_minus'] = False  #마이너스 기호 깨짐 방지
plt.title('Temperature change graph on my birthday')
plt.plot(low_result, 'blue', label='low')
plt.plot(high_result, 'pink', label='high')
plt.legend()
plt.show()
```

##### (3) 1월과 8월의 최고 기온 데이터 히스토그램으로 표현하기

###### (2) 1월과 8월의 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))

#plt.boxplot(jan)
#plt.boxplot(aug)
plt.boxplot([jan, aug])
plt.show()
```

###### (4) 8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)  #>>['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']

aug = []
day = []  #일별 데이터를 저장할 리스트 day 생성

for i in range(31):
    day.append([])  #day 리스트 내에 31개 리스트 생성

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))  #최고 기온 값 저장

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)  #dpi: 선명도
plt.boxplot(day, showfliers=False)
plt.show()
```

##### (5) 1년 중 일교차가 가장 큰 달은 몇 월이고, 몇 도 일까?
- 2019년으로 설정

```python
import csv 
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)  #>>['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']

month = [] 
max_diff = -999
max_month = ''

for row in data:
    if row[0].split('-')[0] == '2019':
        if row[-1] != '' and row[-2] != '':
            month = float(row[-1]) - float(row[-2])  #최고기온-최저기온
            if month > max_diff:
                max_diff = month
                max_month = row[0].split('-')[1]
                
        
print('1년 중 일교차가 가장 큰 달은', max_month,'월이고, 그 날의 일교차는', month, '\'c 였습니다.')
```
