###### 3) 제주도의 성별 인구 비율 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

male_sum = 0
female_sum = 0
male = []
female = []

for row in data:
    if '제주특별자치도' in row[0]:
        for m in row[3:104]:
            male.append(int(m))
        for f in row[106:]:
            female.append(int(f))
        for i in range(101):
            male_sum = male_sum + male[i]
            female_sum = female_sum + female[i]
        break
            
print(male_sum)
print(female_sum)
```

- 제주도의 연령대별 성별 비율을 산점도로 표현하기

```python
#가로축:연령대별 남성 인구, 세로축:연령대별 여성 인구
import csv
import numpy as np
import math
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

male = []
female = []
size = []
region = input('궁금한 동네를 입력해주세요: ')

for row in data:
    if region in row[0]:
        for i in range(3, 104):
            male.append(int(row[i]))
            female.append(int(row[i+103]))
            size.append(math.sqrt(int(row[i]) + int(row[i+103])))
        break
            
plt.style.use('ggplot')
plt.title(region + '지역의 성별 인구 그래프')
plt.rc('font', family='AppleGothic')
plt.scatter(male, female, c=range(101), s=size, alpha=0.75, cmap='jet')
plt.colorbar()
plt.xlabel('male')
plt.ylabel('female')
plt.plot(range(max(male)), range(max(male)), 'g')
plt.show()
```
