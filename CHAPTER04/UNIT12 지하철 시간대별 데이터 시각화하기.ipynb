```python
#7시~9시까지 승차 인원 합친 후 오름차순으로 막대그래프 그리기
import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

result = []

for row in data:
    row[4:] = map(int, row[4:])
    #result.append(row[10]+row[12]+row[14])
    result.append(sum(row[10:15:2]))  #10부터 14까지 2간격으로 = 인덱스 10, 12, 14
    
# print(len(result))
# print(result)
result.sort()
plt.bar(range(len(result)), result)
plt.show()
```

```python
#3시간(7시~9시)동안 사람이 가장 많이 들어오는 역은 어딜까? ->‘최댓값 찾기’
import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

max_result = 0  #최댓값 저장 변수 초기화
max_station = ''  #최댓값을 찾는 역 이름 저장 변수 초기화

for row in data:  #전부 탐색하여 최댓값 갱신
    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > max_result:
        max_result = sum(row[10:15:2])
        max_station = row[3] + '(' + row[1] + ')'

print(max_station, max_result)
```

```python
#출근시간대에 사람들이 가장 많이 내리는 역은 어디일까?(하차)
import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

max_result = 0
max_station = ''

for row in data:
    row[4:] = map(int, row[4:])
    tof = sum(row[11:16:2])  #중복되는 부분 변수 생성하여 코드 최소화하기
    if tof > max_result:
        max_result = tof
        max_station = row[3] + '('+row[1]+')'
        
print(max_station, max_result)
```

###### 3) 밤 11시에 사람들이 가장 많이 타는 역은 어디일까

```python
#밤 11시 - 23:00:00~23:59:59
import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

max_result = 0
max_station = ''

time = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요?(24시 단위): '))

for row in data:
    row[4:] = map(int, row[4:])
    t = 4 + (time - 4) * 2 
    ton = row[t]
    #한줄로 줄이기 -> t = row[4+(time-4)*2]
    
    if ton > max_result:
        max_result = ton
        max_station = row[3] + '('+row[1]+')'
    
print(max_station, max_result)
```

###### 4) 시간대별로 사람들이 가장 많이 타고 내리는 역은 어디일까?

```python
#시간대별 승차 인원이 가장 많은 곳
import csv

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)  #'사용월','호선명','역ID','지하철역','04:00:00~04:59:59','', etc

max_station = ['']*24  #시간대별 최대 승차 인원 역 이름 저장 리스트 초기화
max_result = [0]*24  #시간대별 최대 승차 인원 저장 리스트 초기화

for row in data:
    row[4:] = map(int, row[4:])  #정수 변환
    for i in range(24):
        a = row[4+2*i]  #i와 인덱스 번호 사이 관계식 사용
        if a > max_result[i]:
            max_result[i] = a
            max_station[i] = row[3]

print(max_station)
print(max_result)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

max_result = [0]*24
max_station = ['']*24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[4+2*j]
        if a > max_result[j]:
            max_result[j] = a
            #오전 4시~다음날 오전 3시
            max_station[j] = f'{row[3]}({(j+4)%24}시)'
    
plt.style.use('ggplot')
plt.rc('font', family='AppleGothic')
plt.bar(range(24), max_result)
plt.xticks(range(24), max_station, rotation=90)  #x축 label 90도 회전
plt.show()
```

###### 5) 모든 지하철역에서 시간대별 승하차 인원을 모두 더하면

```python
import csv
import matplotlib.pyplot as plt

f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)

ton_sum = [0]*24  #시간대별 승차 인원 누적 합 리스트 초기화
tof_sum = [0]*24  #시간대별 하차 인원 누적 합 리스트 초기화

for row in data:
    row[4:] = map(int, row[4:])  #승/하차 인원 문자열->정수 변환
    for j in range(24):
        ton_sum[j] += row[4+2*j]  #승차 인원 누적 합
        tof_sum[j] += row[5+2*j]  #하차 인원 누적 합

plt.style.use('ggplot')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.rc('font', family='AppleGothic')
plt.plot(ton_sum, label='승차')  #승차 누적 인원 꺾은선 그래프
plt.plot(tof_sum, label='하차')  #하차 누적 인원 꺾은선 그래프
plt.xticks(range(24), range(4, 28))  #24개로 나누고, 값은 시간대(4시~28시)
plt.legend()
plt.show()
```
