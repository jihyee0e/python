```python
import numpy as np
import matplotlib.pyplot as plt

weight1 = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
        80, 59, 67, 81, 69, 73, 69, 74, 70, 65]
weight2 = [52, 67, 84, 66, 58, 78, 71, 57, 76, 62, 51, 79,
        69, 64, 76, 57, 63, 53, 79, 64, 50, 61]

plt.subplot(2,2,1)
plt.hist((weight1, weight2), histtype='bar')
plt.title('histtype-bar')
#plt.figure()

plt.subplot(2,2,2)
plt.hist((weight1, weight2), histtype='barstacked')
plt.title('histtype-barstacked')
plt.figure()

plt.subplot(2,2,3)
plt.hist((weight1, weight2), histtype='step')
plt.title('histtype-step')
#plt.figure()

plt.subplot(2,2,4)
plt.hist((weight1, weight2), histtype='stepfilled')
plt.title('histtype-stepfilled')
plt.show()
```

```python
data = np.arange(-5, 5)

print(data)
print(data<0)
print(data[data<0])

print()

mask1 = abs(data) > 3 #3보다 큰 절댓값
print(data[mask1])

mask2 = abs(data) % 2 == 0
print(data[mask2])

print()

print(data[mask1|mask2])  # = mask1+mask2 >> [-5 -4 -2 0 2 4] 
print(data[mask1&mask2])  # = mask1*mask2 >> [-4 4]
```

```python
##교재 예제) -50~50 사이가 비어있는 버블 차트 그리기
import numpy as np
import matplotlib.pyplot as plt

#1000개의 랜덤값 추출
x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
size = np.random.randint(10) * 100

mask1 = abs(x) > 50  #x에 저장된 값 중 절댓값이 50보다 큰 값 걸러내기
mask2 = abs(y) > 50  #y에 저장된 값 중 절댓값이 50보다 큰 값 걸러내기

#print(bubble_x[mask1_x])
x = x[mask1+mask2]
y = y[mask1+mask2]

plt.title("mask-bubble chart")
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.75)
plt.colorbar()
plt.show()
```
